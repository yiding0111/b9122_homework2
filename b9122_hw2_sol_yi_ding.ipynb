```python
from bs4 import BeautifulSoup
import requests
from urllib.parse import urljoin
import time
from requests.packages.urllib3.exceptions import InsecureRequestWarning
from selenium import webdriver
from selenium.webdriver.common.by import By
```

# Question 1

```python
def get_press_releases(seed_url, target_count):
    press_releases = []
    visited_urls = set()
    urls_to_visit = [seed_url]

    while urls_to_visit and len(press_releases) < target_count:
        current_url = urls_to_visit.pop(0)
        if current_url in visited_urls:
            continue
        visited_urls.add(current_url)

        response = requests.get(current_url)
        if response.status_code != 200:
            print(f"Failed to retrieve {current_url}")
            continue
        
        soup = BeautifulSoup(response.content, 'html.parser')

        press_release_tag = soup.find('a', {'hreflang': 'en', 'href': '/en/press-release'})
        if press_release_tag:
            if 'crisis' in soup.text.lower():
                press_releases.append(current_url)
                print(f"Found {len(press_releases)} press releases so far...")

        for a_tag in soup.find_all('a', href=True):
            url = urljoin(seed_url, a_tag['href'])
            urls_to_visit.append(url)

    return press_releases

seed_url = 'https://press.un.org/en'

press_releases = get_press_releases(seed_url, target_count=10)

for i, press_release in enumerate(press_releases, 1):
    print(f'Press Release {i}: {press_release}')
```

# Question 2

```python
def fetch_page(url):
    response = requests.get(url)
    return BeautifulSoup(response.text, 'html.parser')

def extract_press_releases(soup):
    press_releases = []
    for article in soup.find_all('article'):
        if article.find('span', class_='ep_name', string='Plenary session'):
            title = article.find('h3').text
            link = article.find('a')['href']
            summary = article.find('p').text
            if 'crisis' in title.lower() or 'crisis' in summary.lower():
                press_releases.append({'title': title, 'link': link, 'summary': summary})
    return press_releases

def main():
    press_releases = []
    page_number = 0

    while len(press_releases) < 10:
        url = f'https://www.europarl.europa.eu/news/en/press-room/page/{page_number}'
        soup = fetch_page(url)
        press_releases += extract_press_releases(soup)
        page_number += 1
    
    for idx, press_release in enumerate(press_releases[:10], start=1):
        print(f"{idx}. Title: {press_release['title']}")
        print(f"   Link: {press_release['link']}")
        print(f"   Summary: {press_release['summary']}\n")

main()
```
